```python
# return saturation vapor pressure
# using Clausius-Clapeyron equation
def sat_p(t):
    return 0.611 * cp.exp(17.67 * (t - 273.15) * ((t - 29.65) ** (-1)))
```

```python
%%time
es = xr.apply_ufunc(sat_p, da, output_dtypes=[float]).rename("saturation_vapor_pressure")
print(es.cupy.is_cupy)
```

```python
# 使用 fsspec 操作 S3 文件系統的示例
import fsspec

# 初始化 S3 文件系統
fs = fsspec.filesystem('s3')

# 列出 S3 存儲桶中的文件
bucket_name = 'your-bucket-name'
files = fs.ls(bucket_name)
print(files)

# 讀取 S3 上的文件
file_path = f'{bucket_name}/path/to/your/file.txt'
with fs.open(file_path, 'r') as f:
    content = f.read()
    print(content)

# 寫入文件到 S3
new_file_path = f'{bucket_name}/path/to/your/new_file.txt'
with fs.open(new_file_path, 'w') as f:
    f.write('Hello, S3!')
print(f'File written to {new_file_path}')
```
